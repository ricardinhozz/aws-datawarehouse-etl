```python
import pandas as pd
import requests
```

```python
url = f'https://api.weatherbit.io/v2.0/history/hourly?city=SaoPaulo&start_date=2025-11-18&end_date=2025-11-19&key=3361e8eba2044015857b77de8707c200'
response = requests.get(url)
response.raise_for_status()
data = response.json()
```

```python

hourly_data = data.get("data", [])

if not hourly_data:
    print("No hourly data found!")
    df = pd.DataFrame()
else:
    df = pd.DataFrame(hourly_data)
    
    columns_to_keep = [col for col in ["datetime", "app_temp", "temp", "weather"] if col in df.columns]
    df = df[columns_to_keep]
    
    if "weather" in df.columns:
        weather_df = pd.json_normalize(df["weather"])
        df = df.drop(columns=["weather"]).join(weather_df)
    
    top_level_cols = {
        "city": data.get("city_name"),
        "lat": data.get("lat"),
        "lon": data.get("lon"),
        "country_code": data.get("country_code"),
        "timezone": data.get("timezone")
    }
    
    for k, v in top_level_cols.items():
        df[k] = v
    
    cols_order = ["city", "lat", "lon", "country_code", "datetime", "temp", 
                  "description", "code", "timezone"]
    df = df[[c for c in cols_order if c in df.columns]]

df
api1 = df.copy()
```

```python
import pandas as pd

def transform_data_from_api_2(api_data_2):
    hourly = api_data_2.get("hourly", {})
    times = hourly.get("time", [])
    
    if not times:
        return pd.DataFrame()  # no data
    
    # Create DataFrame from hourly lists
    df = pd.DataFrame(hourly)
    
    # Rename columns to match first API
    df = df.rename(columns={
        "temperature_2m": "temp",
        "cloud_cover": "code",  # you can adjust if you want a different name
        "time": "datetime"
    })
    
    # Add top-level info
    df["lat"] = api_data_2.get("latitude")
    df["lon"] = api_data_2.get("longitude")
    df["city"] = api_data_2.get("city_name", None)  # city_name might not exist
    df["country_code"] = api_data_2.get("country_code", None)
    df["timezone"] = api_data_2.get("timezone")
    
    # Optionally, add description if needed (not in this API)
    if "description" not in df.columns:
        df["description"] = None
    
    # Reorder columns
    cols_order = ["city", "lat", "lon", "country_code", "datetime", "temp", 
                  "description",  "timezone"]
    df = df[[c for c in cols_order if c in df.columns]]
    df['city'] = 'São Paulo'
    return df


api2 = transform_data_from_api_2(data)
```
